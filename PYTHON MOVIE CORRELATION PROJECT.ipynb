```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configuration of the plots we will create


# Read in the data 

df = pd.read_csv(r'/Users/michael.vidal/Downloads/movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our colums

df.dtypes
```

```python
# Change data type of colums

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)

df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 
```

```python
df.head()
```

```python
# Create correct Year column

df['release_year'] = df['released'].astype(str).str.split().str[2]

df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Display max rows

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df_dups = df['company'].drop_duplicates().sort_values(ascending=False)

df_dups.head()
```

```python
df.head()
```

```python
# Scatter plot with budget vs gross


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vd gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "orange"}, line_kws={"color":"blue"})
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
        
df_numerized.head()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr.head()
```
